```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('/content/drive/My Drive/data_train.csv')
test_data = pd.read_csv('/content/drive/My Drive/data_test.csv')
val_data = pd.read_csv('/content/drive/My Drive/data_val.csv')
```

```python
X_train = pd.read_csv('/content/drive/My Drive/X_train_w2v.csv').values
X_val   = pd.read_csv('/content/drive/My Drive/X_val_w2v.csv').values
X_test  = pd.read_csv('/content/drive/My Drive/X_test_w2v.csv').values
```

```python
# целевая переменная
y_train = train_data['target']
y_val = val_data['target']
y_test = test_data['target']
```

```python
train_data['target'].value_counts()
```

Данные несбалансированы -> берем метрику f1



```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, f1_score
```

### LogisticRegression

```python
# создание модели
clf_lr = LogisticRegression(
        penalty = 'l2',
        C = 0.2,
        class_weight = 'balanced',
        random_state = 42,
        max_iter = 100,
)
```

```python
# обучение
clf_lr_pretrained = clf_lr.fit(X_train, y_train)
```

```python
clf_lr_pretrained
```

```python
# предсказания
y_predict_val = clf_lr_pretrained.predict(X_val)
y_predict_test = clf_lr_pretrained.predict(X_test)
```

### RandomForestClassifier

```python
# создание модели
clf_rf =  RandomForestClassifier(
        n_estimators = 200,
        criterion = 'gini',
        max_depth = 200,
        class_weight = 'balanced',
        random_state = 42
)
```

```python
# обучение
clf_rf = clf_rf.fit(X_train, y_train)
# предсказание
y_predict_val_rf = clf_rf.predict(X_val)
y_predict_test_rf = clf_rf.predict(X_test)
```

### Метод опорных векторов

```python
# создание модели
clf_lsvc= LinearSVC(
        penalty = 'l2',
        C = 0.2,
        class_weight = 'balanced',
        random_state = 42,
        max_iter = 2000
)
```

```python
# обучение
clf_lsvc = clf_lsvc.fit(X_train, y_train)
# предсказания
y_predict_val_lsvc = clf_lsvc.predict(X_val)
y_predict_test_lsvc = clf_lsvc.predict(X_test)
```

### Оценка результатов

```python
f1_val_lr = f1_score(y_val, y_predict_val)
f1_test_lr = f1_score(y_test, y_predict_test)
```

```python
f1_val_rf = f1_score(y_val, y_predict_val_rf)
f1_test_rf = f1_score(y_test, y_predict_test_rf)
```

```python
f1_val_lsvc = f1_score(y_val, y_predict_val_lsvc)
f1_test_lsvc = f1_score(y_test, y_predict_test_lsvc)
```

```python
results = pd.DataFrame({
    'Классификатор': ['LogisticRegression', 'RandomForest', 'LinearSVC'],
    'F1_val': [f1_val_lr, f1_val_rf, f1_val_lsvc],
    'F1_test': [f1_test_lr, f1_test_rf, f1_test_lsvc]
})

print(results)
```

```python
print("F1 (binary):", f1_score(y_val, y_predict_val, average='binary'))
print("F1 (macro):", f1_score(y_val, y_predict_val, average='macro'))
print("F1 (weighted):", f1_score(y_val, y_predict_val, average='weighted'))
print("F1 (micro):", f1_score(y_val, y_predict_val, average='micro'))
```

Наилучшие результаты на валидации и тесте показывает модель RandomForest
